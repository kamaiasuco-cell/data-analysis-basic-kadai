```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib


df = pd.read_csv("sample_pandas_6.csv")

category_df = pd.read_csv('category.csv')

# dfとcategory_dfを商品番号で結合して、dfにカテゴリー列を追加
# 冗長な2回目のマージを削除し、列名の衝突によるリネームを防ぎます。
# 元のdfの全行を保持するためにhow='left'を使用します。
df = pd.merge(df, category_df[['商品番号', 'カテゴリー']], how='left', on='商品番号')

# カテゴリー列の出現頻度をカウントして棒グラフで表示
df["カテゴリー"].value_counts().plot(kind="bar")

plt.title("カテゴリー別の出現頻度")
plt.xlabel("カテゴリー")
plt.ylabel("件数")
plt.show()

# 商品番号ごとに注文数の基本統計量を求める
result = df.groupby("商品番号")["注文数"].describe()

print(result)
```
