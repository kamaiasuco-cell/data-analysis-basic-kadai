```python
import matplotlib.pyplot as plt
import numpy as np

# 2行3列のサブプロットを作成
fig, axes = plt.subplots(2, 3, tight_layout=True)

# 1. 折れ線グラフ
y0 = [1, 2, -5, 2]
axes[0, 0].plot(y0)
axes[0, 0].set_title("Line Plot")

# 2. sin関数
x1 = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x1)
axes[0, 1].plot(x1, y1)
axes[0, 1].set_title("sin(x)")

# 3. ヒストグラム
data2 = np.random.randn(100)
axes[0, 2].hist(data2)
axes[0, 2].set_title("Histogram")

# 4. 散布図
x3 = np.random.rand(100)
y3 = np.random.rand(100)
axes[1, 0].scatter(x3, y3)
axes[1, 0].set_title("Scatter Plot")

# 残り2領域は何も書かなくてOK
axes[1, 1].axis("off")
axes[1, 2].axis("off")

# グラフを表示
plt.show()
```
